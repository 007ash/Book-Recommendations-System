### Rating Data Set

```python
ratings = pd.read_csv("data/BX-Book-Ratings.csv", sep=";", encoding="latin-1", on_bad_lines='skip')
```

```python
ratings['user_id'].unique().shape
```

### Analysis  and Preprocessing of Data Set

```python
# how many users have rated more than 200 books
X = ratings['user_id'].value_counts() > 200
print(X)  # ithu True or False value la varum
print(X[X].shape)
print(X[X].index) 
```

```python
# True data value mattu vanum
y = X[X].index 
ratings = ratings[ratings['user_id'].isin(y)]
print(ratings.shape)
ratings
```

#### Merge the Rating and Books

```python
print(books.columns)
print(ratings.columns)
```

```python
# ISBN same ah irruku so atha vaichu merge pannalam (DBMS la join pannura mathiri)
# mela join panninaalum, ratings la irukura books ellam books la irukkum
ratings_books = ratings.merge(books, on='ISBN')
```

```python
ratings_books.columns
```

```python
print(ratings_books.shape)
ratings_books.head()
```

```python
# How many times each book has been rated?
num_ratings = ratings_books.groupby('title')['rating'].count().reset_index()
num_ratings.rename(columns={'rating': 'num_of_ratings'}, inplace=True)
num_ratings.head()
```

#### Merging the num_rating and ratings_books

```python
print(ratings_books.columns)
print(num_ratings.columns)
```

```python
final_ratings = ratings_books.merge(num_ratings, on='title')
```

```python
print(final_ratings.columns)
print(final_ratings.shape)
```

```python
# 50 ratings ku athigam irukura books mattu venum
# ithu tha ratings and book final DataFrame
final_ratings = final_ratings[final_ratings['num_of_ratings'] >= 50]
print(final_ratings.shape)
final_ratings.head()
```

```python
final_ratings.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_ratings.shape
```

#### Pivot table

```python
book_pivot = final_ratings.pivot_table(index='title', columns='user_id', values='rating')
```

```python
print(book_pivot.shape)
book_pivot.head()
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot.head()
```

```python
# 0 athigam ah irrutha clustering model slow ah work aagum
# so atha avoid pannalaam athuku csr_matrix use pannuvom (M3 la paathom la) 
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot.values) 
```

### Model Building

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distance, suggestions = model.kneighbors(book_pivot.iloc[27, :].values.reshape(1, -1), n_neighbors=6)
```

```python
print(distance)
print(suggestions)
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
book_names = book_pivot.index
book_names
```

### Model Saving

```python
import numpy as np
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index ==  book_name)[0][0]
    distance, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    for i in range(len(suggestions)):
        books = book_pivot.index[suggestions[i]]
        for j in books:
            print(j) 
```
